```python
import pandas as pd
```

```python
df=pd.read_csv('agora.csv')
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
marketing_scale_fit=scaler.fit(df[['Marketing Spend']])
```

```python
marketing_scale_fit
```

```python
df['Marketing Spend']=scaler.transform(df[['Marketing Spend']])
```

```python
df
```

```python
df['Administration']=scaler.fit_transform(df[['Administration']])
```

```python
df
```
